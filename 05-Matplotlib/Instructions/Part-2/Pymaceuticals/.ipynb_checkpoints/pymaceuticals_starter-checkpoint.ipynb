```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
mouse_drug_df = pd.read_csv(mouse_drug_data_to_load)
clinical_trial_df = pd.read_csv(clinical_trial_data_to_load)



# Combine the data into a single dataset

combined_data_df = pd.merge(clinical_trial_df, mouse_drug_df, how="left", on= "Mouse ID")
# Display the data table for preview

combined_data_df.head()
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 

tumor_mean = combined_data_df.groupby(["Drug","Timepoint"]).mean() ["Tumor Volume (mm3)"]

# Convert to DataFrame
tumor_mean_df = pd.DataFrame(tumor_mean)

# Preview DataFrame

tumor_mean_df.reset_index().head()
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint

tumor_se = combined_data_df.groupby(["Drug","Timepoint"]).sem() ["Tumor Volume (mm3)"]

# Convert to DataFrame

tumor_se_df = pd.DataFrame(tumor_se)

# Preview DataFrame

tumor_se_df.reset_index().head()
```

```python
# Minor Data Munging to Re-Format the Data Frames

# Preview that Reformatting worked
```

```python
tumor_mean_df = tumor_mean_df.reset_index()
tumor_pivot_mean = tumor_mean_df.pivot(index="Timepoint",columns = "Drug")["Tumor Volume (mm3)"]

tumor_pivot_mean
```

```python
tumor_se_df = tumor_se_df.reset_index()
tumor_pivotse_df = tumor_se_df.pivot(index="Timepoint",columns = "Drug")["Tumor Volume (mm3)"]

tumor_pivotse_df
```

```python
# Generate the Plot (with Error Bars)

plt.errorbar(tumor_pivot_mean.index, tumor_pivot_mean["Capomulin"],yerr=tumor_pivotse_df["Capomulin"],color="r", marker="o", markersize=5, linestyle="dashed", linewidth=0.50, )
plt.errorbar(tumor_pivot_mean.index, tumor_pivot_mean["Infubinol"],yerr=tumor_pivotse_df["Infubinol"],color="b", marker="^", markersize=5, linestyle="dashed", linewidth=0.50)
plt.errorbar(tumor_pivot_mean.index, tumor_pivot_mean["Ketapril"],yerr=tumor_pivotse_df["Ketapril"],color="g", marker="s", markersize=5, linestyle="dashed", linewidth=0.50)
plt.errorbar(tumor_pivot_mean.index, tumor_pivot_mean["Placebo"],yerr=tumor_pivotse_df["Placebo"],color="k", marker="d", markersize=5, linestyle="dashed", linewidth=0.50)


plt.title("Tumor Response to Treatment")
plt.xlabel("Tumor Volume 3mm")
plt.ylabel("Time in Days")
plt.grid()
plt.show()
# Save the Figure

```

## Metastatic Response to Treatment

```python
# Store the Mean Met. Site Data Grouped by Drug and Timepoint 
metastatic_mean = combined_data_df.groupby(["Drug","Timepoint"]).mean() ["Metastatic Sites"]
metastatic_mean_df = pd.DataFrame(metastatic_mean)
metastatic_mean_df.head()
# Convert to DataFrame

# Preview DataFrame
```

```python
# Store the Standard Error associated with Met. Sites Grouped by Drug and Timepoint 
metastatic_se = combined_data_df.groupby(["Drug","Timepoint"]).sem() ["Metastatic Sites"]
# Convert to DataFrame
metastatic_se_df = pd.DataFrame(metastatic_se)
# Preview DataFrame
metastatic_se_df.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
metastatic_mean_df = metastatic_mean_df.reset_index()
metastatic_pivot_mean = metastatic_mean_df.pivot(index="Timepoint",columns = "Drug")["Metastatic Sites"]

metastatic_pivot_mean 

# Preview that Reformatting worked
```

```python
metastatic_se_df = metastatic_se_df.reset_index()
metastatic_pivot_se = metastatic_se_df.pivot(index="Timepoint",columns = "Drug")["Metastatic Sites"]

metastatic_pivot_se
```

```python
# Generate the Plot (with Error Bars)
plt.errorbar(metastatic_pivot_mean.index, metastatic_pivot_mean["Capomulin"],yerr=metastatic_pivot_se["Capomulin"],color="r", marker="o", markersize=5, linestyle="dashed", linewidth=0.50)
plt.errorbar(metastatic_pivot_mean.index, metastatic_pivot_mean["Infubinol"],yerr=metastatic_pivot_se["Infubinol"],color="b", marker="^", markersize=5, linestyle="dashed", linewidth=0.50)
plt.errorbar(metastatic_pivot_mean.index, metastatic_pivot_mean["Ketapril"],yerr=metastatic_pivot_se["Ketapril"],color="g", marker="s", markersize=5, linestyle="dashed", linewidth=0.50)
plt.errorbar(metastatic_pivot_mean.index, metastatic_pivot_mean["Placebo"],yerr=metastatic_pivot_se["Placebo"],color="k", marker="d", markersize=5, linestyle="dashed", linewidth=0.50)

plt.title("Metastatic spread during treatment")
plt.xlabel("Met. Sites")
plt.ylabel("Treatment Duration (days)")
plt.grid()
plt.show()
# Save the Figure

# Show the Figure
```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)

survival_count = combined_data_df.groupby(["Drug","Timepoint"]).count()["Tumor Volume (mm3)"]

# Convert to DataFrame

survival_count_df = pd.DataFrame({"Mouse Count": survival_count})


# Preview DataFrame
survival_count_df.reset_index().head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
survival_count_df = survival_count_df.reset_index()
survival_count_pivot = survival_count_df.pivot(index="Timepoint",columns = "Drug")["Mouse Count"]
# Preview the Data Frame
survival_count_pivot 
```

```python
# Generate the Plot (Accounting for percentages)
plt.plot(100 * survival_count_pivot["Capomulin"]/25,"ro", color="r", marker="o", markersize=5, linestyle="dashed", linewidth=0.50)
plt.plot(100 * survival_count_pivot["Infubinol"]/25,"ro", color="b", marker="^",linestyle="dashed", markersize=5, linewidth=0.50) 
plt.plot(100 * survival_count_pivot["Ketapril"]/25,"ro", color="g", marker="s",linestyle="dashed", markersize=5, linewidth=0.50) 
plt.plot(100 * survival_count_pivot["Placebo"]/25,"ro",color="k", marker="d", linestyle="dashed", markersize=5, linewidth=0.50) 

plt.title("Survival During Treatment")
plt.xlabel("Survival Rate %")
plt.ylabel("Time(days)")
plt.grid()
plt.show()

# Save the Figure

# Show the Figure
plt.show()
```
