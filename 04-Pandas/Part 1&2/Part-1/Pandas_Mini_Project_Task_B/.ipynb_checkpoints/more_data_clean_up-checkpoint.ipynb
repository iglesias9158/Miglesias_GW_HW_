```python
# Dependencies
import pandas as pd
```

```python
# load CSV
csv_path = "Resources/2016-FCC-New-Coders-Survey-Data.csv"
```

```python
# Read with pandas--low_memory required to suppress errors about mixed data types
surveydata_df = pd.read_csv(csv_path)
```

```python
# Take only columns 0, 1, 2, 3, 4, 6, 7, 8, 9, 10, 11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111

reduced_more_clean = surveydata_df.iloc[:,[0, 1, 2, 3, 4, 7, 8, 9, 10, 11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111]]
```

```python
# Change reading 0 and 1 to No and Yes, respectively

reduced_more_clean = reduced_more_clean.replace ({0.0: "No", 1.0:"Yes"})
```

```python
# Extract rows for only those who attended a bootcamp

more_attended = reduced_more_clean.loc[reduced_more_clean["AttendedBootcamp"] == "Yes"]
```

```python
# Create DataFrame of the different boot camps that had a significant number of attendees

att_bootcamp = more_attended.groupby(["BootcampName_"]).count()["AttendedBootcamp"]
bootcamp_data_attend = pd.DataFrame({"AttendedBootcamp": att_bootcamp })
bootcamp_data_attend.head(10)
```

```python
# Count how many attendees of each bootcamp would recommend the bootcamp

rec_bootcamp = more_attended.groupby(["BootcampName_"]).count()["BootcampRecommend"].rename("Recommends the Bootcamp")
bootcamp_data_recommend = pd.DataFrame({"Recommends the Bootcamp": rec_bootcamp })
bootcamp_data_recommend
```

```python
# Merge the two created data frames on the name of tbe bootcamp

merge_table = pd.merge(bootcamp_data_attend, bootcamp_data_recommend, on="BootcampName_")
merge_table.head(10)
```

```python
# Calculate percentage of each bootcamp's students who are recommenders
merge_table['Result'] = (merge_table['Recommends the Bootcamp']/merge_table['AttendedBootcamp']).map("{0:,.2f}%".format)
merge_table.head(10)
```

```python
# Export to excel and remove index
merge_table.to_excel("output/bootcamppart2.xlsx", index=False)
```
